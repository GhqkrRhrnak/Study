```python
# import

import pandas as pd
```

### Sereis

```python
# dictionary data
data = {"a" : 1, "b" : 2, "c" : 3}
print(data)
print(type(data))
print()

# Series
sr = pd.Series(data)
print(sr)
print(type(sr))
print()

idx = sr.index
val = sr.values
print(idx)
print(val)
print()
print()
```

```python
# tuple data
data = ('영인', '2010-05-01', '여', True)
sr = pd.Series(data, index = ['이름', '생년월일', '성별', '학생여부'])
print(sr)
print()

print(sr[0])
print(sr['이름'])
print()

print(sr[['이름', '성별']])
print(sr[[1, 3]])
print()

print(sr[1:3]) # range
print(sr['이름':'성별'])
```

### DataFrame

```python
data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}

df = pd.DataFrame(data)
display(df)
print(type(df))
print()

df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']],
                 index = ['준서', '예은'],
                 columns = ['나이', '성별', '학교'])
display(df)

df.index = ['학생1','학생2']
df.columns = ['연령', '남녀', '소속']
display(df)

df.rename(index = {'학생1':'준서', '학생2':'예은'}, 
         columns = {'연령':'나이', '남녀':'성별', '소속':'학교'}, 
         inplace = True)
display(df)
```

```python
# drop
df2 = df[:]
df2.drop('예은', inplace = True)

df3 = df[:]
df3.drop(['나이', '성별'], axis = 1, inplace = True)

display(df)
display(df2)
display(df3)
```

```python
# location
# loc :    a:d -> a~b
# iloc :   0:3 -> 0~2
print(df.loc['준서':'예은'])
print(df.iloc[0:1])
print()

print(df.loc['준서':]) # 준서 ~ 끝
print(df.iloc[:1])     # 처음 ~ 지정인덱스-1 
print(df.iloc[:-1])    # 처음 ~ 끝-1
print()
# -1 : 인덱스 가장 끝

print(df.loc[['준서', '예은']])
print(df.iloc[[0, 1]])
print()
```

```python
# 열선택
print(df.나이)
print(df['성별'])
print()

print(df[['나이', '학교']])
print(df.iloc[:, 0:2])
print(df.loc[:, "나이":"학교"])
print()
```

```python
# 열 데이터 추가
df['키'] = 160 
# display(df)
# 일괄 적용. 모든 row의 키 데이터가 160으로 들어감

# 행 데이터 추가
df.loc['학생3'] = [17, '여', '덕영중', 160]
display(df)
```

```python
# 데이터 변경

# 인덱스를 학교로 변경
# 기존에 있던 인덱스는 사라짐
df.set_index('학교', inplace = True)
display(df)

df.loc['덕영중']['나이'] = 16
df.loc['수리중', '나이'] = 15
display(df)


# 행 열 위치 변경
df = df.T
display(df)
df = df.transpose()
display(df)
```

```python
# # 행 인덱스 재배열
# new_index = ['a', 'b', 'c']
# df = df.reindex(new_index, fill_value = 0)
# # 인덱스를 새로 지정하고, NaN 값(누락값)은 0으로 채워넣음
# display(df)
# 인덱스이름 들어간 상태로 재배열 불가능함.. 왜지 -> 중복 column 존재시 발생하는 오류
# df.index

# 행 인덱스 초기화
# 인덱스가 정수형 위치 인덱스로 변경됨
# 기존에 있던 인덱스는 사라지지 않고 밀려남
df = df.reset_index()
display(df)

# 행 인덱스 재배열
df.index = ['학생1', '학생2', '학생3']
display(df)
# 기존에 있던 인덱스가 새로운 인덱스로 바뀌면 누락값으로 채워짐
new_index = ['학생1', '학생2', '학생3', '학생4', '학생5']
df = df.reindex(new_index, fill_value = 0)
# 인덱스를 새로 지정하고, NaN 값(누락값)은 0으로 채워넣음
display(df)

# 인덱스 정렬
# default : asc = True 
df = df.sort_index()
display(df)
df = df.sort_index(ascending = False)
display(df)
```

```python
# 열 데이터 기준 정렬

df.sort_values(by='키', ascending = False)
# 문자는 대소 비교 할 수 없기 때문에 정렬 불가
```

### 연산

```python
sr1 = pd.Series({'국어':70, '영어':80, '수학':90})
sr2 = pd.Series({'국어':40, '수학':60})
# print(sr1, sr2)

# NaN과 연산하면 NaN이 나옴
print(sr1 + sr2) # NaN

sr3 = sr1.add(sr2, fill_value = 0)
# sr1 + sr2. NaN값은 0으로 채움
print(sr3)
```
